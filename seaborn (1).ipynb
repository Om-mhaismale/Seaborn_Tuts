***HEAT MAP - works only on 2D data(Array)***

***Box plot***

```python
box=sns.load_dataset("tips")
sns.set(style="whitegrid")
box
```

```python
sns.boxplot(x="day",y="total_bill",hue="time",data=box,palette="plasma",showmeans=True,meanprops={"marker":"+","markeredgecolor":"black"},linewidth=5,fill=False,order=["Sun","Sat","Fri","Thur"])
# meanprops to change properties of mean
# orient parameter is only possible for numeric value, but if you want u can simply swap x,y axis
```

1. Box: Shows the middle 50% of your data.
2. Line inside the box: Represents the median (middle value).
3. Whiskers(-------|): Extend from the box to show the overall range of the data, excluding outliers(dots).
4. Dots outside the whiskers: Indicate outliers, or data points that are much higher or lower     than the rest.

```python
sns.boxplot(data=box,palette="plasma",showmeans=True,meanprops={"marker":"+","markeredgecolor":"black"},linewidth=2,fill=True,orient="h")
```

```python
sns.boxplot(y=box["total_bill"])
mt.show()
```
